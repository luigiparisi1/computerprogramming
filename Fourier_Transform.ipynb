```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_components = 5
gains = np.random.randn(N_components) #generate 5 random gains
sr = 4000
T = 2
Nsamples = sr*T
time = np.linspace(0,T, Nsamples)
x = np.zeros(time.shape)
F0 = 1
for i in range(N_components):
  F=F0*(50*i)+1
  c= gains[i]*np.cos(2*np.pi*time*F)
  plt.figure()
  plt.plot(time,c)
  plt.show()
  x = x + c
```

```python
plt.plot(time, x)
```

```python
from scipy import fftpack
```

```python
X = fftpack.fft(x)
```

```python
X
print(x.shape, X.shape ) #uno nel time domain, uno nel frequency domain
```

```python
plt.plot(np.abs(X)) # 5 peaks for the components and nothing else, symmetric
plt.show()
```

```python
X1 = np.fft.fft(x)
print(X[:10])
print(X1[:10])
```

```python
plt.plot(np.abs(X)/Nsamples)
```
